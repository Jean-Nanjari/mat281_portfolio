1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
x = df[['month_number']]
x1 = df[['facecream']]
x2 = df[['facewash']]
x3 = df[['toothpaste']]
x4 = df[['bathingsoap']]
x5 = df[['shampoo']]
x6 = df[['moisturizer']]
x7 = df[['total_units']]
y = df['total_profit']
```

```python

plt.plot(
    x, # eje x
    y, # eje y
    label='linea', # etiquetado
    color="black", # color
    linewidth=1 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.title("mes v/s ganancia") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('ganancia') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

```python
plt.scatter(
    x, # eje x
    y, # eje y
    label='linea', # etiquetado
    color="black", # color
    linewidth=0.5 # tamano de la curva
)
```
